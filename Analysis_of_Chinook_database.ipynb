```python
%%sql
with info as 
(select count(quantity) as number_tracks_sold ,genre.name as genre_name from invoice 
inner join invoice_line on invoice.invoice_id=invoice_line.invoice_id 
inner join track on invoice_line.track_id=track.track_id 
inner join genre on genre.genre_id=track.genre_id
where invoice.billing_country= "USA"
group by genre.genre_id
order by number_tracks_sold desc)


select number_tracks_sold, 
round(cast(number_tracks_sold as float)/(select sum(number_tracks_sold) from info)*100,2)
as number_tracks_sold_perc,
genre_name 
from info 
group by genre_name
order by number_tracks_sold_perc desc
```

```python
%%sql
select customer.country,count(distinct(customer.customer_id)) as total_customers,
sum(invoice.total) as total_salves_val,
sum(invoice.total)/count(distinct(customer.customer_id)) as avg_sales_per_cus,
avg(invoice.total) as avg_ord_val
from customer
inner join invoice on customer.customer_id=invoice.customer_id
group by country
having count(distinct(customer.customer_id))>1
order by count(distinct(customer.customer_id)) desc
```

```python
%%sql
select "Other" as country, sum(total_customers) as total_customers, 
sum(total_salves_val)as total_salves_val, 
SUM(avg_sales_per_cus)/sum(total_customers) as avg_sales_per_cus,
SUM(avg_ord_val)/sum(total_customers) as avg_ord_val
from
(select customer.country as country, count(distinct(customer.customer_id)) as 
 total_customers,
sum(invoice.total) as total_salves_val,
sum(invoice.total)/count(distinct(customer.customer_id)) as avg_sales_per_cus,
avg(invoice.total) as avg_ord_val
from customer inner join invoice on 
customer.customer_id=invoice.customer_id
group by customer.country
having count(distinct(customer.customer_id))=1)
```

```python
%%sql
with details_by_country as 
(
select customer.country,count(distinct(customer.customer_id)) as total_customers,
sum(invoice.total) as total_salves_val,
sum(invoice.total)/count(distinct(customer.customer_id)) as avg_sales_per_cus,
avg(invoice.total) as avg_ord_val
from customer
inner join invoice on customer.customer_id=invoice.customer_id
group by country
having count(distinct(customer.customer_id))>1
    
union 

select "Other" as country, sum(total_customers) as total_customers, 
sum(total_salves_val)as total_salves_val, 
SUM(avg_sales_per_cus)/sum(total_customers) as avg_sales_per_cus,
SUM(avg_ord_val)/sum(total_customers) as avg_ord_val
from
(select customer.country as country, count(distinct(customer.customer_id)) as 
 total_customers,
sum(invoice.total) as total_salves_val,
sum(invoice.total)/count(distinct(customer.customer_id)) as avg_sales_per_cus,
avg(invoice.total) as avg_ord_val
from customer inner join invoice on 
customer.customer_id=invoice.customer_id
group by customer.country
having count(distinct(customer.customer_id))=1)    
)

select * from details_by_country
order by total_customers
```

```python
%%sql
with album_track_count as 
(select invoice.invoice_id,invoice_line.invoice_line_id,invoice_line.track_id,
count(distinct(track.album_id)) as album_count
from invoice inner join invoice_line 
on invoice.invoice_id=invoice_line.invoice_id 
inner join track on track.track_id=invoice_line.track_id
group by invoice.invoice_id)

select count(*) as album_purchase from album_track_count
where album_count=1
```

```python
%%sql
with album_track_count as 
(select invoice.invoice_id,invoice_line.invoice_line_id,invoice_line.track_id,
count(distinct(track.album_id)) as album_count
from invoice inner join invoice_line 
on invoice.invoice_id=invoice_line.invoice_id 
inner join track on track.track_id=invoice_line.track_id
group by invoice.invoice_id)

select count(*) as track_purchase from album_track_count
where album_count!=1
```

```python
%%sql 
create view purchased_tracks as 
select count(distinct(invoice_line.track_id)) as tracks_by_genre_purchased,
genre.name as genre_name
from invoice_line inner join track on invoice_line.track_id=track.track_id
inner join genre on track.genre_id=genre.genre_id
group by genre.genre_id
```

These are the number of tracks purchased by customers. 
Now let us look at tracks by genre present in the store.

```python
%%sql 
create view all_tracks as
select count(distinct(track.track_id)) number_of_tracks, 
genre.name as genre_name from track 
inner join genre on track.genre_id=genre.genre_id
group by genre.name 
```

```python
%%sql 
select all_tracks.genre_name, all_tracks.number_of_tracks as total_tracks_in_store,
purchased_tracks.tracks_by_genre_purchased as tracks_purchased,
round(cast(purchased_tracks.tracks_by_genre_purchased as float)/all_tracks.number_of_tracks*100,2) as percentage_unsold
from all_tracks inner join purchased_tracks 
on all_tracks.genre_name=purchased_tracks.genre_name
order by percentage_unsold
```

- The table above very clearly informs us that tracks in the Drama and TV Shows genre almost go unsold. So if managemnet was looking to remove tracks of any particular genre these 2 would be the best. 
- In terms of number Rock, Metal, Alternative & Punk ans Latin sell the most. 
- In terms of highest percentage of tracks sold, Easy Listening Electronica/Dance, R&B Soul sells the best, they almost sell all the copies

```python
%%sql

select 
"Protected" as type,
count(track.track_id) as count_tracks
from media_type inner join track 
on media_type.media_type_id=track.media_type_id
where media_type.name like "%Protected%"

union 
select "Non-Protected" as type, 
count(track.track_id) from media_type 
inner join track on media_type.media_type_id=track.media_type_id 
where media_type.name not like "%Protected%"
```
